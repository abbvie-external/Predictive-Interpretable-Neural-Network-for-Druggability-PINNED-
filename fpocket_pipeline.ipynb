```python
def begin(url ):
    url_output = pd.read_csv(url, sep= ':')
    #print(url_output)
    url_output.columns = ['Properties', 'Pocket']
    url_pocket = url_output['Pocket']
    
    new = url_output['Properties'].str.split('\t',expand = True, n =1)
    
    tab = new[0]
    Properties = new[1]
    Properties = Properties[0:19]
    
    protein_info = define_pockets(url_pocket, Properties, )  
    print(protein_info)
    return protein_info 
    
```

```python
def drop_na(protein):
    protein = protein.dropna()
    protein = protein.reset_index(drop = True)
    
    return protein
```

```python
def adjust_header(protein):
    protein = protein.rename(columns = protein.iloc[0]).drop(protein.index[0])
    
    return protein
```

```python
frames = []
def define_pockets(protein, Properties):
    
    protein =  drop_na(protein)
    a = 0
    b = 19
    len_protein =  len(protein)//19+1
    names = ['Properties']
    Pocket_all = []
    Pocket_all.append(Properties)
    
    for i in range (1, len_protein):
        Pocket = []
        Pocket = protein[a:b].reset_index(drop = True)
        Pocket_all.append(Pocket)
        a+=19
        b+=19
        #print(Pocket_all)
       

    for j in range (1, len_protein):
        names.append('Pocket '+ str(j))
    new_protein = pd.concat(Pocket_all, keys = names ,axis=1)
    
    new_protein = new_protein.T
      
    new_protein =  adjust_header(new_protein)
    master_list = new_protein 
    print(master_list)   
    merged = find_drug_scores(new_protein, master_list,  Properties)
     
    frames.append(merged)
    result = pd.concat(frames)
    #print(type(result))
    return result
```
